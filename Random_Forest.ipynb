```python
import pandas as pd
```

```python
df = pd.read_csv('/content/temps.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
X_train = pd.get_dummies(df)
```

```python
X_train.iloc[:,5:]
```

```python
X_train.iloc[:,0:].head()
```

```python
import numpy as np
```

```python
labels = np.array(X_train['actual'])

X_train = X_train.drop('actual', axis=1)
```

```python
feature_list = list(X_train.columns)
```

```python
feature_list
```

```python
X_train.head()
```

```python
features = np.array(X_train)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(features, labels, test_size=0.25, random_state=42)
```

```python
train_x.shape, train_y.shape,  test_x.shape, test_y.shape
```

```python
baseline_preds = test_x[:,feature_list.index('average')]
baseline_preds
```

```python
baseline_errors = abs(baseline_preds-test_y)
round(np.mean(baseline_errors),2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train_x, train_y)
```

```python
predictions = rf.predict(test_x)
```

```python
errors = abs(predictions - test_y)

round(np.mean(errors), 2)
```

```python
mape = 100 * (errors/test_y)
accuracy = 100 - np.mean(mape)
accuracy
```

```python
rf_small = RandomForestRegressor(n_estimators=10, max_depth=3)
rf_small.fit(train_x, train_y)

tree_small = rf_small.estimators_[5]
```

```python
importances = list(rf.feature_importances_)
importances
```

```python
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]

feature_importances = sorted(feature_importances, key = lambda x: x[1] , reverse = True)

feature_importances
```
